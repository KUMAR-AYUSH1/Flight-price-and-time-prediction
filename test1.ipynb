```python
df.drop(['ch_code','num_code'],axis=1,inplace=True)
```

```python
df['time'] =  df['dep_time'].str.split(":").str[0].astype(int)
```

```python
def time_of_day(x):
    if x >= 5 and x < 9:
        return "early morning"
    elif x >= 9 and x < 12:
        return "morning"
    elif x >= 12 and x < 16:
        return "afternoon"
    elif x >= 16 and x < 20:
        return "evening"
    else:
        return "night"
df['time_of_day'] = df['time'].apply(time_of_day)
```

```python
df.drop(['time'],axis=1,inplace=True)
```

```python
df['is_weekday'] = df['date'].dt.day_of_week
df['month'] = df['date'].dt.month

df['is_weekday'].value_counts()
```

```python
df["time_taken"].value_counts()
```

```python
df['hrs'] = df['time_taken'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['min'] = df['time_taken'].str.extract(r'(\d+)m').fillna(0).astype(int)

df['total_duration_mins'] = (df['hrs'] * 60) + df['min']

df.drop(['hrs', 'min','time_taken'],axis=1,inplace=True)

print(df['total_duration_mins'].describe())
```

```python
df.drop(['dep_time','arr_time'],axis=1,inplace=True)
```
