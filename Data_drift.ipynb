```python
import pandas as pd
import numpy as np
import joblib
from scipy.stats import chi2_contingency
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
old_data = pd.read_csv('clean_data.csv')
new_data = pd.read_csv('test.csv')

old_data.drop(['date'], axis=1, inplace=True)
new_data.drop(['date'], axis=1, inplace=True)

x = new_data.drop(['total_duration_mins','price'],axis=1)
y= new_data[['total_duration_mins','price']]
```

```python
ct = joblib.load('ct.joblib')
RandomForestRegressor_MultiOutputRegressor_model = joblib.load('RandomForestRegressor_MultiOutputRegressor_model.joblib')
RandomForestRegressor_RegressorChain_model = joblib.load('RandomForestRegressor_RegressorChain_model.joblib')
XGBRegressor_MultiOutputRegressor_model = joblib.load('XGBRegressor_MultiOutputRegressor_model.joblib')
XGBRegressor_RegressorChain_model = joblib.load('XGBRegressor_RegressorChain_model.joblib')

x = ct.transform(x)
```

```python
models = [RandomForestRegressor_MultiOutputRegressor_model,RandomForestRegressor_RegressorChain_model,XGBRegressor_MultiOutputRegressor_model,XGBRegressor_RegressorChain_model]

for model in models:
    predictions = model.predict(x)
    print(f"{model}\n")
    print(f"R2 Score: {r2_score(y, predictions)}")
    print(f"Mean Absolute Error: {mean_absolute_error(y, predictions)}")
    print(f"Mean Squared Error: {mean_squared_error(y, predictions)}")
    print(f"Root Mean Squared Error: {root_mean_squared_error(y, predictions)}")
    print("\n")
```

```python
# Data drift check



cols = ['airline', 'from', 'stop', 'to', 'class', 'time_of_day','is_weekday', 'month']
ALPHA = 0.05
for TARGET_COLUMN in cols:
    old_counts =old_data[TARGET_COLUMN].value_counts()
    new_counts = new_data[TARGET_COLUMN].value_counts()
    value_count_table = pd.DataFrame({'old': old_counts,'new': new_counts}).fillna(0)

    print("=== value_count Table ===")
    print(value_count_table, "\n")

    chi2, p_val, dof, expected = chi2_contingency(value_count_table, correction=True)
    print(f"total size of the gap  {chi2}, probability  of drift {p_val},  \n")
    if p_val < ALPHA:
        print(f"\n[ALERT] Data Drift Detected! The distribution has changed significantly (p < {ALPHA}).")
    else:
        print(f"\n[OK] No significant drift detected (p >= {ALPHA}).")
```
